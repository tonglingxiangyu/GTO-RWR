# t-SNE Visualization
**Here we present the t-SNE plots of all the four algorithms, E-GraphSAGE and its modified version, GAT and E-ResGAT, on four datasets. The plots are based on a batch of 5000 testing records for each dataset.**

## UNSW-NB15

```python
from sklearn.manifold import TSNE
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
test = np.load("UNSW-NB15_test.npy")
X_1 = np.load("UNSW-NB15_emb_1.npy")
X_2 = np.load("UNSW-NB15_emb_2.npy")
X_3 = np.load("UNSW-NB15_emb_3.npy")
X_4 = np.load("UNSW-NB15_emb_4.npy")
X = [X_1, X_2, X_3, X_4]
y = np.load("UNSW-NB15 - CSV Files/label_mul.npy")
palette = sns.color_palette("bright", 10)

X_embed = []
for i in range(4):
    tsne = TSNE(random_state=42, perplexity=20)
    X_embed.append(tsne.fit_transform(X[i]))

import matplotlib.pyplot as plt
fig, axs = plt.subplots(nrows=2,ncols=2, figsize=(12,10))
axs[0][0].set_xlabel('(a)', fontsize=16)
axs[0][1].set_xlabel('(b)', fontsize=16)
axs[1][0].set_xlabel('(c)', fontsize=16)
axs[1][1].set_xlabel('(d)', fontsize=16)
sns.scatterplot(X_embed[0][:,0], X_embed[0][:,1], hue=y[test], legend='brief', palette = palette, ax=axs[0][0])
sns.scatterplot(X_embed[1][:,0], X_embed[1][:,1], hue=y[test], legend=False, palette = palette, ax=axs[0][1])
sns.scatterplot(X_embed[2][:,0], X_embed[2][:,1], hue=y[test], legend=False, palette = palette, ax=axs[1][0])
sns.scatterplot(X_embed[3][:,0], X_embed[3][:,1], hue=y[test], legend=False, palette = palette, ax=axs[1][1])
plt.show()
```

```python
test = np.load("UNSW-NB15_test.npy")
X_1 = np.load("UNSW-NB15_emb_1.npy")
X_2 = np.load("UNSW-NB15_emb_2.npy")
X_3 = np.load("UNSW-NB15_emb_3.npy")
X_4 = np.load("UNSW-NB15_emb_4.npy")
X = [X_1, X_2, X_3, X_4]
y = np.load("UNSW-NB15 - CSV Files/label_mul.npy")
id_ = np.where(y[test]!=0)[0]
id_ = np.delete(id_, np.argmax(X_embed[0][id_,0]))
idx = np.delete(id_, np.argmax(X_embed[0][id_,0]))
palette = sns.color_palette("bright", 10)[1:]

X_embed = []
for i in range(4):
    tsne = TSNE(random_state=42, perplexity=20)
    X_embed.append(tsne.fit_transform(X[i]))

import matplotlib.pyplot as plt
fig, axs = plt.subplots(nrows=2,ncols=2, figsize=(12,10))
axs[0][0].set_xlabel('(a)', fontsize=20)
axs[0][1].set_xlabel('(b)', fontsize=20)
axs[1][0].set_xlabel('(c)', fontsize=20)
axs[1][1].set_xlabel('(d)', fontsize=20)
sns.scatterplot(X_embed[0][idx,0], X_embed[0][idx,1], hue=y[test][idx], legend='brief', palette = palette, ax=axs[0][0])
sns.scatterplot(X_embed[1][idx,0], X_embed[1][idx,1], hue=y[test][idx], legend=False, palette = palette, ax=axs[0][1])
sns.scatterplot(X_embed[2][idx,0], X_embed[2][idx,1], hue=y[test][idx], legend=False, palette = palette, ax=axs[1][0])
sns.scatterplot(X_embed[3][idx,0], X_embed[3][idx,1], hue=y[test][idx], legend=False, palette = palette, ax=axs[1][1])
plt.tight_layout(pad=0)
plt.savefig('tsne-unsw.png', dpi=300)
```
